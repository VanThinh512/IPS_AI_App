```python
import pandas as pd
import numpy as np

df = pd.read_csv('tuesday_balanced_label_training_1.csv', sep =',')

df.columns = (df.columns.str.strip())

for i, col in enumerate(df.columns):
    print(i, repr(col))

```

```python
# chọn lọc đặc trưng (feature selection)
# list of optimal columns for brute force

selected_features = [
    #--time group--
    'flow duration',
    'flow iat mean', 'flow iat std', 'flow iat max', 'flow iat min',
    'fwd iat mean', 'fwd iat std',

    #--size/count group--
    'total fwd packets', 'total backward packets',
    'fwd packet length mean', 'fwd packet length std',
    'bwd packet length mean', 'bwd packet length std',
    'flow packets/s', 'flow bytes/s',

    #--flags/windows group--
    'fwd psh flags', 'ack flag count', 'syn flag count' , 'fin flag count',
    'init_win_bytes_forward', 'init_win_bytes_backward',

    #label
    'label'
]

#only include the columns in the list above 
df_reduced = df[selected_features].copy()

```

```python
#làm sạch dữ liệu
#thay thế inf = NaN
df_reduced.replace([np.inf, -np.inf], np.nan, inplace = True)
df_reduced.dropna(inplace = True)
print(f"Kích thước ban đầu: {df.shape}")
print(f"Kích thước sau khi lọc cột: {df_reduced.shape}")
print("Danh sách cột được chọn:")
print(df_reduced.columns.tolist())

print(df_reduced.head())
```

```python
df_reduced.to_csv('reduced_tuesday_bruteforce_columns_2.csv', index=False)
```
